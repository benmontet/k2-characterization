```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re

from isochrones.starmodel import StarModel
from isochrones.dartmouth import Dartmouth_Isochrone
from isochrones.padova import Padova_Isochrone

DARTMOUTH = Dartmouth_Isochrone()
PADOVA = Padova_Isochrone()

import k2fpp
```

```python
epic_id = 201208431
mags = k2fpp.mags(epic_id)
exclude_bands = ['W1','W2','W3','W4'] #don't have WISE mags in isochrones
exclude_bands += ['B','V'] #Padova grids don't have
```

```python
kws = {}
for k,v in mags.items():
    if re.search('err',k):
        continue
    if k not in exclude_bands:
        kws[k] = (v, mags['{}err'.format(k)])
model = StarModel(DARTMOUTH, maxAV=0.1, **kws)
model_pad = StarModel(PADOVA, maxAV=0.1, **kws)
```

```python
#model_pad.fit_mcmc()
model.fit_mcmc()
```

```python
hist_kws = {'lw':3, 'histtype':'step'}
model.plot_samples('Teff', **hist_kws)
#model_pad.plot_samples('Teff',fig=0, **hist_kws)
model.plot_samples('mass', **hist_kws)
#model_pad.plot_samples('mass', fig=0, **hist_kws)
model.plot_samples('feh', **hist_kws)
#model_pad.plot_samples('feh', fig=0, **hist_kws)
model.plot_samples('distance', **hist_kws)
#model_pad.plot_samples('distance', fig=0, **hist_kws)
model.plot_samples('AV', **hist_kws)
#model_pad.plot_samples('AV', fig=0, **hist_kws)
```
