```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import re

from isochrones.starmodel import StarModel
from isochrones.dartmouth import Dartmouth_Isochrone

DARTMOUTH = Dartmouth_Isochrone()

import k2fpp
```

```python
epic_id = 201208431
mags = k2fpp.mags(epic_id)
exclude_bands = ['W1','W2','W3','W4'] #don't have WISE mags in isochrones
```

```python
kws = {}
for k,v in mags.items():
    if re.search('err',k):
        continue
    if k not in exclude_bands:
        kws[k] = (v, mags['{}err'.format(k)])
model = StarModel(DARTMOUTH, maxAV=0.4, **kws)
```

```python
model.fit_mcmc()
```

```python
hist_kws = {'lw':3, 'histtype':'step'}
model.plot_samples('Teff', **hist_kws)
model.plot_samples('mass', **hist_kws)
model.plot_samples('feh', **hist_kws)
model.plot_samples('distance', **hist_kws)
model.plot_samples('AV', **hist_kws)
```
